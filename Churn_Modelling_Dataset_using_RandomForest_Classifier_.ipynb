```python
from sklearn.preprocessing import LabelEncoder
f=LabelEncoder()
for i in df:
  df[i]=df[[i]].apply(f.fit_transform)
```

```python
df.drop(['Gender','Age'],axis=1)
```

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a=StandardScaler()
a.fit_transform(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(random_state=70,n_jobs=-1,max_depth=10,n_estimators=150,max_features=20,oob_score=True,criterion='entropy',max_leaf_nodes=40)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
